```python
df_price_drop = df.drop(['Price'],axis = 1).copy()
```

```python
categorical_data = df.select_dtypes(include='object')
categorical_data.head()
```

```python
df1 = df.copy()
df1["Area"] = pd.cut(df1['Area'], bins =[0,250,500,750,1000,1250,1500,1750,2000,2250,2500,2750,3000])
fig = plt.figure(figsize=(30,50))
for i,j in enumerate(df_price_drop):
    fig.add_subplot(10,4,i+1)
    sns.countplot(df1[j])
    plt.xlabel(j,size=20)
    plt.xticks(rotation=90)

plt.show()
```

```python
fig,stack_bar= plt.subplots(figsize=(30,45))
stack_bar.set_title('House at Each Location in Mumbai, Maharastra', fontsize=50)
value_location_index = df.Location.value_counts().index[:30]
sns.countplot(y="Location",data=df, order=value_location_index)
stack_bar.set_xlabel('Locations', fontsize = 30)
stack_bar.set_ylabel('Number of Houses', fontsize = 30)
plt.show()
```

```python
fig = plt.figure(figsize=(18,200))
sns.countplot(df1['Area'])
plt.xlabel('Area', fontsize= 30)
plt.ylabel('Number of Houses', fontsize= 30)
plt.xticks(rotation=40)
plt.show()
```

```python
df2 = df.copy().replace(9,np.nan)
df2 = df2.fillna(method='bfill',axis=0).fillna(0)
df2.head()
```

```python
correlation = df2.select_dtypes(exclude='object')
num_correlation =correlation.corr()
plt.figure(figsize=(20,20))
plt.title('Correlation of Mumbai DataSet', fontsize=50)
sns.heatmap(num_correlation>0.8, annot =True, square = True)
```

```python
num_correlation['Price'].sort_values(ascending = False)
```
